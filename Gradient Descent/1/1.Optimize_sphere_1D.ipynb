# Sphere Function and Its Derivative

The sphere function is defined as:

$f(x) = x^2$


The derivative of the sphere function is:

$f'(x) = 2x$

```python
# sphere function

def func(x):
    return x**2

def derivative(x):     
    return 2*x
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-100, 100, 0.1)
y = func(x)

plt.plot(x, y)
plt.show()
```


# Optimization Process

```python
import random

num_steps = 100
lr = 0.1

# 1. set x randomly
x = random.randint(-100, 100)
x_scatter = [x]

print(x)

for _ in range(num_steps):
    # compute derivative
    dx = derivative(x)
        
    # update
    x = x - lr*dx
    x_scatter.append(x)
    print(x, func(x))
```

# Visualization of Optimization Process

```python
x = np.arange(-100, 100, 0.1)
y = func(x)

for i in range(len(x_scatter)):
    plt.scatter(x_scatter[i], func(x_scatter[i]))

plt.plot(x, y)
plt.show()
```
